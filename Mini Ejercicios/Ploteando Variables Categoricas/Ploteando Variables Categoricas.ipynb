## Ploteando masivamente variables categoricas

En este ejercicio corto realizaré un analisis (muuy rapido) de como plotear variables categorícas de forma masiva.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings #Para eliminar l os warnings

# load in the data
df = pd.read_csv("mushroom_data.csv")
print(df.head())
```

```python
df.dtypes
```

Son todas object!, sumando que Bruises es de tipo boolean, habria que cambiarlo a string u object para no tener problemas mas adelante con otras librerias.

```python
df['Bruises'] = df['Bruises'].apply(lambda x: str(x))
df.dtypes
```

En resumen, son 23 variables, todas string, por lo que este dataset es 100% variables categoricas. Para preparar el ejercicio extraemos la lista de columnas a la variable "columns"

```python
# Lista de todas las columnas
columns = df.columns.tolist()
```

Procedemos a iterar cada columna. En el caso de que la variable tenga 3 o menos categorías ploteamos un "pie" chart, solo para poner a prueba mis if y elif jeje.

```python
for i in columns:
    if len(df[i].unique()) <=3: #Si la variable contiene 3 o menos categorias ploteamos un pie chart
        plt.pie(df[i].value_counts(), labels=df[i].unique().tolist(),
                autopct="%.1f%%")
    elif len(df[i].unique()) >= 4: #Si  contiene  4 o más ploteamos un grafico de barra ordenado por cantidad
        sns.countplot(df[i], order=df[i].value_counts().index)
        plt.xticks(rotation = 30, fontsize=12) #Amononamos un poquito
    plt.title(i)
    plt.show()
    plt.clf()

warnings.filterwarnings('ignore')
```
