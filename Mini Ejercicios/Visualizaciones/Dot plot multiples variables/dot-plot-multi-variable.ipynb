```python
# Creación de un dot plot multi variable con apuntes
# Tutorial : https://seaborn.pydata.org/examples/pairgrid_dotplot.html

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# Cargando dataset
df = sns.load_dataset("car_crashes")

# Haciendo el PairGrid
g = sns.PairGrid(df.sort_values("total",ascending=False), # El dataset ordenado por "total".
                x_vars = df.columns[:-3], # Queremos como variables eje x a todas las columnas hasta el indice -3
                y_vars = ["abbrev"], # Nuestro eje y seran las ciudades abreviadas!
                height = 10, aspect=.25 # Pequeñas configuraciones como altura y aspecto
)
# Explicación mas profunda del PairGrid : https://www.youtube.com/watch?v=Skyp6-RmA84

# Dibujamos un dot plot usando la funcion stripplot
# .map me permite plotear cualquier grafico en el grid
g.map(sns.stripplot, size = 10, orient = "h", jitter = False,
    palette = "flare_r",linewidth = 1, edgecolor = "w") # stripplot genera un scatterplot separado por categorias

# Usamos el mismo eje x en todas las columnas y agregamos un label
g.set(xlim=(0,25), xlabel="Crashes",ylabel="") # Seteamos xlim, xlabel y ylabel

# Agregamos titulos a cada subplot (Lo hago despues de ver como queda el plot)
titulos = ["Total de accidentes", "Acc. Exceso velocidad","Accidentes por alcohol", "Acc. sin distracción",
            "Sin accidentes anteriores"]
for ax,titulo in zip(g.axes.flat, titulos):

    # Setiamos un titulo diferente para cada eje
    ax.set(title=titulo)
    
    # Grid horizontal
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True) # Remueve las grids, en este caso la de la izq y la del fondo
plt.show()
```

```python
df2 = pd.read_csv("C:/Users/chuck/Documents/Ejemplos Python/2018w48.csv")
df2.head(15)

# Resumen del dataset: Es una comparación de precios 
```

```python
df2.item.unique()
# Al parecer si queremos hacer un Grid parecido al de arriba, tenemos que pivotear la tabla, en donde cada item sea una categoria!
# Porque? porque quiero plotear cuanto cuesta cada item por ciudad, (ignoraremos entonces : category, country y sort item)
df_pivot = df2.pivot_table(index = "city", columns = "item", values="cost", aggfunc="sum",margins=True).sort_values("city",ascending=True).drop("All",axis=1).reset_index()
# Se ve bien, tenemos precios de cada item por ciudad!, pero tenemos "city y All"
# Resumen, hicimos un pivto table por indice ciudad, columnas item, valores costo, sumamos todos y se nos generó una columna Total que por la cual ordenamos y despues dropeamos
# Luego resteamos el indice para que quede asi.
df_pivot = df_pivot.drop(0) # Dropeamos el "All" que nos creo el pivot
df_pivot
```

```python
df_pivot["2 Longdrinks"] = df_pivot["2 Longdrinks"].str.replace(",",".").astype(float)
df_pivot.info()

# Nos dejo todos los numeros como objeto!!!, y otro problema, al hacer la pivot table nos asigno una , y python para poder transformar a numero debe ser un "."
```

```python
# Ahora, a todos
df_pivot["Big Mac"] = df_pivot["Big Mac"].str.replace(",",".").astype(float)
df_pivot["Cinema entry"] = df_pivot["Cinema entry"].str.replace(",",".").astype(float)
df_pivot["Club entry"] = df_pivot["Club entry"].str.replace(",",".").astype(float)
df_pivot["Dinner"] = df_pivot["Dinner"].str.replace(",",".").astype(float)
df_pivot["Drinks at Dinner"] = df_pivot["Drinks at Dinner"].str.replace(",",".").astype(float)
df_pivot["Taxi"] = df_pivot["Taxi"].str.replace(",",".").astype(float)
```

## Extra: Como aplicar la funcion a columnas que eran numeros "object" con "," y hacer el procedimiento de arriba, pero masivamente?

```python
# Con un map podemos hacer esto, las funciones ya las tenemos, pero queremos especificar que columnas queriamos cambiar

columnas_malas = ["Big Mac","Cinema entry", "Club entry", "Dinner", "Drinks at Dinner", "Taxi","2 Longdrinks"]

# Creamos una funcion

def transformar(numero):
    return numero.replace(",",".")

# Probemoslo
print(transformar("13,2"))
```

```python
# Aplicamos el map a las columnas seleccionadas
# Recreemos todo denuevo

df_pivot_test = df2.pivot_table(index = "city", columns = "item", values="cost", aggfunc="sum",margins=True).drop("All",axis=1).reset_index()
# Se ve bien, tenemos precios de cada item por ciudad!, pero tenemos "city y All"
# Resumen, hicimos un pivto table por indice ciudad, columnas item, valores costo, sumamos todos y se nos generó una columna Total que por la cual ordenamos y despues dropeamos
# Luego resteamos el indice para que quede asi.
df_pivot_test = df_pivot_test.drop(12)

df_pivot_test[columnas_malas] = df_pivot_test[columnas_malas].applymap(transformar)
df_pivot_test

```
